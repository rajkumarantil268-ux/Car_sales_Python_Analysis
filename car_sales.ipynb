```python
df.groupby("Category")["Price"].agg(
    ["sum","mean","max","min"]
)
```

```python
df.sort_values(
    by="Price",
    ascending=False
)
```

```python
pd.pivot_table(
    df,
    values="Price",
    index="Category",
    aggfunc="sum"
)
```

```python
import matplotlib.pyplot as plt
Total_Price = df.groupby(
    "Category"
)["Price"].sum()

Total_Price.plot(
    kind="bar"
)

plt.title("Revenue by Industry")
plt.show()
```

```python
Years = df.groupby(
    "Years"
)["Price"].sum()

Years.plot(
    kind="line"
)

plt.show()
```

```python
Total_Price.plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.show()
```
